```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

matplotlib.interactive(True)
plt.ion()
matplotlib.is_interactive()
```

```python
# There are several features that can be setted up
plt.plot([2,3,4,5],[4,6,8,10])
```

```python
# Color
plt.plot([2,3,4,5],[4,6,8,10], color = 'green')
```

```python
plt.plot([2,3,4,5],[4,6,8,10], color = 'green')
# Labels
plt.xlabel ('X', fontsize = 16, color = 'g')
plt.ylabel ('2*X', fontsize = 16, color = 'g')
```

```python
# If not provided values for the X axis the values 
# displayed will be indices for each value provided
plt.plot ([4,8,12,16])
```

```python
# There can be plotted Math functions given a set of X
# values
x = np.linspace(start = 0, stop = 10, num = 50)
plt.plot(x,np.sin(x), color = 'green')
plt.xlabel ('X', fontsize = 16, color = 'g')
plt.ylabel ('SIN(X)', fontsize = 16, color = 'g')
```

```python
# Tick parameters can be setted under the following method
# Y axis
plt.tick_params(axis='y', color = 'red', labelcolor = 'b', labelsize = 'xx-large')
plt.plot(x,np.sin(x), color = 'green')
```

```python
# X axis
plt.tick_params(axis='x', color = 'b', labelcolor = 'green', labelsize = 'xx-large')
plt.plot(x,np.sin(x), color = 'green')
```

```python
# Line labels
plt.plot(x,np.sin(x), label = 'Sine Curve label')
plt.title('Sine Curve title')
plt.legend()
```

```python
# Limits
# X axis
plt.plot(x,np.sin(x))
plt.xlim(1,5)
```

```python
# Y axis
plt.plot(x,np.sin(x))
plt.ylim(-0.5,0.5)
```

```python
# Invert over an axis
# The function goes from a smaller to a greated value over
# an axis, if the limit follows the inverse disposition
# it will mirror the plot over the limited axis
# X axis
plt.plot(x,np.sin(x))
plt.xlim(5,1)
```

```python
plt.plot(x,np.sin(x))
plt.xlim(1,5)
```

```python
# Y axis
plt.plot(x,np.sin(x))
plt.ylim(-1,1)
```

```python
plt.plot(x,np.sin(x))
plt.ylim(1,-1)
```
