```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

matplotlib.interactive(True)
plt.ion()
matplotlib.is_interactive()
```

```python
# Scales
# Any plot has its own scales
# By default the scale is linear
np.random.seed()
y_1 = np.random.uniform(low=0.0, high=100, size=(1000,))
y_1.sort()

x_1 = np.arange(len(y_1))
plt.plot(x_1,y_1)
plt.grid(True)
plt.yscale('linear') # Linear scale of the Y axis
```

```python
plt.plot(x_1,y_1)
plt.yscale('log') # Log scale of the Y axis
```

```python
plt.plot(x_1,y_1)
plt.xscale('log') # Linear scale of the X axis
```

```python
plt.plot(x_1,y_1)
plt.yscale('log',base = np.exp(1))
```
