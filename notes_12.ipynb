```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

matplotlib.interactive(True)
plt.ion()
matplotlib.is_interactive()
```

```python
# Use case of BoxPlots
# Data preprocessing
data_a = pd.read_csv('resources/exams.csv')
data_b = data_a[['math score', 'reading score','writing score']]
data_c = np.array(data_b)

# Plot
plot_a = plt.boxplot(data_c, patch_artist=True)

# Style 
colors = ['red','purple','orange']

# Process
for box in range(len(plot_a['boxes'])):
    plot_a['boxes'][box].set(facecolor=colors[box])
    plot_a['caps'][2 * box + 1].set(color=colors[box])

plt.xticks([1,2,3],['Math score', 'Reading Score', 'Writing score'])
```

```python
# Use case of Violin plots
# Plot
plot_b = plt.violinplot(data_c, showmedians=True)
plt.xticks([1,2,3],['Math score', 'Reading Score', 'Writing score'])
```

```python
# Plot
plot_c = plt.violinplot(data_c, showmedians=True, vert= False)
plt.yticks([1,2,3],['Math score', 'Reading Score', 'Writing score'])
```

```python
# Plot
plot_d = plt.violinplot(data_c, showmedians=True, vert= False)

# Process
for bodies in range(len(plot_d['bodies'])):
    plot_d['bodies'][bodies].set(facecolor=colors[bodies])


plt.legend (handles = [plot_d['bodies'][0], plot_d['bodies'][1], plot_d['bodies'][2]],
            labels = ['Math','Reading','Writing'],
            loc = 'upper left')

plot_d
```

```python
# Use case of Piecharts
# Data preprocessing
data_g = pd.read_csv('resources/sector_weighting.csv')

# Plot
plt.pie(data_g['Percentage'],labels = data_g['Sector'])
```

```python
# There can be provided different colors. The number of declared
# colors is meant to be the same ammount of the sections as the pie charts have
colors = ['lawngreen','orange','teal','violet','yellow']
plt.pie(data_g['Percentage'],labels = data_g['Sector'], colors = colors,autopct = '%.2f')
```

```python
# A Pie chart can be plotted clockwise with a division starting at 12'0
plt.pie(data_g['Percentage'],labels = data_g['Sector'], colors = colors,autopct = '%.2f', startangle=90, counterclock=False)
```

```python
# There's a way to separate a particular set of values from the pie chart
explode_a = (0,0.1,0,0.5,0) 
# Only values at index 1 and 3 will be separated starting clockwise
plt.pie(data_g['Percentage'],labels = data_g['Sector'], colors = colors,autopct = '%.2f', startangle=90, counterclock=False, explode=explode_a)
plt.axis('equal')
```

```python
# There are several values returned from the pie chart
wedges, texts, autotext = plt.pie(data_g['Percentage'],labels = data_g['Sector'], colors = colors,autopct = '%.2f', explode=explode_a)
plt.axis('equal')
# Note that each value stands for:
# 1 - Wedge: Each segment of the piechart
# 2 - Text: Each one of the labels for each segment
# 3 - Autotext: Texts generated by the plot like percentage with the positions

wedges[1].set(edgecolor = 'purple', linewidth = 5)
texts[1].set(family = 'monospace', size = 20)
autotext[1].set(weight = 'bold', size = 25)
```

```python
# Use case of Autocorrelations in Timeseries data

# Autocorrelation is the measure of the relation between a variable's current 
# value and past values. The autocorrelation value can range between -1 and 1,
# being 1 perfect positive autocorrelation and -1 perfect negative autocorrelation

# Data preprocessing
data_h = pd.read_csv('resources/grand_canyon_visits.csv')
# Note: There's a limit of memory usage whenever working with numerical values, 
# mostly integer called 'OverfloError'
# In computing, an overflow error is an error that happens when a program receives
# a number, value or variable outside the scope of its ability to handle. This
# type of error is somewhat common in programming, especially when dealing with
# integers or other numerical types of variables.

# When working with large ammount of data there's a common practice that consists
# in divide the dataset values between a power of 10 to deal with smaller 
# values to avoid overflow
data_h['NumVisits']
```

```python
data_h['NumVisits'] = data_h['NumVisits'] / 1000
data_h['NumVisits']
```

```python
# Autocorrelation requires the study of multiple previous steps. Note that 
# this number of steps can be between 1 and the total number of records - 1
plt.figure(figsize=(16,8))
plt.acorr(data_h['NumVisits'], maxlags = 20)
```

```python
# This graph shows that theres a high relation at 0 meaning that there is exactly
# no difference between a value at 0 steps from itself (duh!). Like comparing
# how a value is different to itself. At any steps from the reference value there 
# will be a difference that, aside from the reference value itself, have particularly
# high correlation (observable in the graph) at 12 steps forward and backwards in 
# time. (This type of correlation indicates seasonality)

# Every autocorrelation plot returns a tuple of values
plt.figure(figsize=(16,8))
lags, c, vlines, hlines = plt.acorr(data_h['NumVisits'], maxlags = 20)
# Note that each value stands for:
# 1 - Lags: List of each lag value in the correlation plot
# 2 - C: List containing each of the correlation values
# 3 - Vlines: List of each patch values for each correlation value
# 4 - Hlines: The horizontal line value
```

```python
# Use case of Stacked Plots
# Data preprocessing
data_j = pd.read_csv('resources/national_parks.csv')
data_k = data_j['Year']
data_l = np.vstack([data_j['Badlands'],
                    data_j['GrandCanyon'],
                    data_j['BryceCanyon']])

labels_a = ['Badlands','GrandCanyon','BryceCanyon']
colors_a = ['180A0A','711A75','F10086']

# Plot
plt.stackplot(data_k,data_l, labels=labels_a, colors=colors_a, edgecolor= 'black')
plt.legend(loc='upper left')
```

```python
# Use case of Stem Plots
# Data preprocessing
data_j[['Badlands','GrandCanyon','BryceCanyon']] = data_j[['Badlands','GrandCanyon','BryceCanyon']].diff()

colors_b = ['F68989','C65D7B','874356']

# Plot
plt.figure(figsize=(10,6))
plt.stem(data_k,data_j['Badlands'])
plt.title('Change in number of visitors')
```

```python
plt.figure(figsize=(10,6))
plt.stem(data_k,data_j['Badlands'],markerfmt= 'r_',linefmt = 'g--', basefmt = 'b:')
plt.title('Change in number of visitors')
```
