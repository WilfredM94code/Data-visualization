Data visualization is a tool used in the field of 'Data analysis' to provide an evidence an observable clues about a phenomena of study, helping to display insights more intuitively.

For such task there has been developed multiple tool for data visualization and among all there is a tool called Matplotlib. 

Matplotlib is a Python package used to create 2D plots given sets of data and it works with  Pandas and Numpy.

Matplotlib is a packages wich integrates:
- Matplotlib - pyplot
- Pylab (collection of commands and functions to make matplotlib to work like MATLAB)
- Matplotlib APIs

Every Matplotlib plot is considered as a data container
that consists (hierarchically) in:
- Figure
- Axes (X and Y)
- Ticks (major and minor)

A figure can cantain also:
- Grid
- Line plot
- Markers

Matplotlib has an interactive and a non-interactive mode.
In the interactive mode, the graph display gets updated after each statement. In the non-interactive mode, the graph does not get displayed until explicitly asked to do so 

```python
import matplotlib

# Interactive mode off
matplotlib.interactive(False)
print (matplotlib.is_interactive())
# Interactive mode on
matplotlib.interactive(True)
print (matplotlib.is_interactive())
```

```python
# This can be also achieved by executing the following 
# syntax
import matplotlib.pyplot as plt
# Interactive mode off
plt.ioff()
print (matplotlib.is_interactive())
# Interactive mode on
plt.ion()
print (matplotlib.is_interactive())
```
