```python
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from matplotlib.path import Path

matplotlib.interactive(True)
plt.ion()
matplotlib.is_interactive()
```

```python
# Beizer curves can be created using Matplotlib
# Bizer curve is a parametric curve used in computer graphics and related fields. 
# A set of discrete "control points" defines a smooth, continuous curve by means of
# a formula.

# This type of curves are used in computer modelig for industries such as
# 1 - Computer graphic, vector graphic
# 2 - Modelling, animation, typography
# 3 - Web development
```

```python
fig_a, ax_a = plt.subplots()
points = patches.PathPatch(Path([(0.1,0.1),(0.8,0.8),(0.8,0.1),(0.4,0.2)],
                                [Path.MOVETO,Path.LINETO,Path.LINETO,Path.LINETO]),
                                fill = None)

ax_a.add_patch (points)
# The Path object consists in Points and the interconnection between these points
# The 'Path.LINETO' object changes the position between points stablishing a line
# The 'Path.MOVETO' object changes the position without stablishing a line
```

```python
# To have a closed polygon there can be used the 'Path.CLOSEPOLY'
# object
fig_b, ax_b = plt.subplots()
points = patches.PathPatch(Path([(0.1,0.1),(0.8,0.8),(0.8,0.1),(0.4,0.2)],
                                [Path.MOVETO,Path.LINETO,Path.LINETO,Path.CLOSEPOLY]))

ax_b.add_patch (points)
```

```python
# From these paths there can be created Bezier curves
fig_c, ax_c = plt.subplots()
points = patches.PathPatch(Path([(0.1,0.1),(0.8,0.8),(0.8,0.1),(0.4,0.2)],
                                [Path.MOVETO,Path.LINETO,Path.CURVE3,Path.CURVE3]))

ax_c.add_patch (points)
# The 'Path.CURVE3' object creates a Bezier curve. This notation implies the usage
# of 3 control points to create a curve, which creates a cuadratic Bezier curve.
# To change move the anchor position/value there can placed an additional value
# in the constructor
```

```python
fig_d, ax_d = plt.subplots()
points = patches.PathPatch(Path([(0.1,0.1),(0.8,0.8),(0.8,0.1),(0.4,0.2),(0.1,0.1)],
                                [Path.MOVETO,Path.LINETO,Path.CURVE4,Path.MOVETO,Path.MOVETO]))

ax_d.add_patch (points)
```
