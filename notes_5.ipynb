```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

matplotlib.interactive(True)
plt.ion()
matplotlib.is_interactive()
```

```python
# Plotting stock data to 
data_a = pd.read_csv('resources/stocks.csv')
data_a
```

```python
data_a['Date'] = pd.to_datetime(data_a['Date'])
data_a.head()
```

```python
fig_a = plt.figure(figsize=(10,10))
ax_a = fig_a.add_axes([0,0,1,1])
ax_b = fig_a.add_axes([0.05,0.65,0.5,0.3])

ax_a.plot(data_a['Date'], data_a['AAPL'], color = 'g')
ax_b.plot(data_a['Date'], data_a['IBM'], color = 'r')
ax_a.set_title('Apple vs IBM')
# Way to plot several stock data frames in a plot
```

```python
fig_b = plt.figure(figsize=(10,10))
fig_b.suptitle('Stock price comparison 2007-2017')
ax_c = fig_b.add_subplot(221)
ax_c.set_title('Microsoft')
ax_c.plot(data_a['Date'], data_a['MSFT'], color = 'g')

ax_c = fig_b.add_subplot(222)
ax_c.set_title('Google')
ax_c.plot(data_a['Date'], data_a['GOOG'], color = 'r')

ax_c = fig_b.add_subplot(223)
ax_c.set_title('Starbucks')
ax_c.plot(data_a['Date'], data_a['SBUX'], color = 'b')

ax_c = fig_b.add_subplot(224)
ax_c.set_title('Chevron')
ax_c.plot(data_a['Date'], data_a['CVX'], color = 'y')
fig_b
```
