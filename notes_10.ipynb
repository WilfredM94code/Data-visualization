```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

matplotlib.interactive(True)
plt.ion()
matplotlib.is_interactive()
```

```python
# Twin axis
data_a = pd.read_csv('resources/austin_weather.csv')
data_b = data_a[['Date','TempAvgF','WindAvgMPH']].head(30)
```

```python
fig_a, ax_temp_a = plt.subplots()
fig_a.set_figwidth(12)
fig_a.set_figheight(6)

ax_temp_a.set_xlabel('Date')
ax_temp_a.tick_params(axis='x', bottom = False, labelbottom=False)

ax_temp_a.set_ylabel('Temperature (F)', color = 'orange', size= 'x-large')
ax_temp_a.tick_params(axis='y', labelcolor = 'orange', labelsize='large')

ax_temp_a.plot(data_b['Date'], data_b['TempAvgF'], color = 'orange')

ax_temp_b = ax_temp_a.twinx()

ax_temp_b.set_ylabel('Average Wind Speed (MPH)', color = 'green', size = 'x-large')
ax_temp_b.tick_params(axis= 'y', labelcolor = 'green', labelsize= 'large')

ax_temp_b.plot(data_b['Date'], data_b['WindAvgMPH'], color = 'green')
```

```python

# Twin axis with diferent scale

def fahrenheit_2_celcius(f):
    return (f - 32) * 5/9

fig, ax_tempF = plt.subplots()

fig.set_figwidth(12)
fig.set_figheight(6)

ax_tempF.set_xlabel("Date")
ax_tempF.tick_params(axis="x", bottom=False, labelbottom=False)

ax_tempF.set_ylabel("Temperature (F)", color="blue", size="x-large")
ax_tempF.tick_params(axis="y", labelcolor="blue", labelsize="large")

ax_tempF.plot(data_b['Date'], data_b['TempAvgF'], color="blue")

ax_tempC = ax_tempF.twinx()

ymin, ymax = ax_tempF.get_ylim()

# Scaled twin axis
ax_tempC.set_ylim(fahrenheit_2_celcius(ymin), fahrenheit_2_celcius(ymax))

ax_tempC.set_ylabel("Temperature (C)", color="purple", size="x-large")
ax_tempC.tick_params(axis="y", labelcolor="purple", labelsize="large")
```
