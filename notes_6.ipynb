```python
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.patches as patches

matplotlib.interactive(True)
plt.ion()
matplotlib.is_interactive()
```

```python
# There are several shapes that can be plotted using the patches module
fig_a, ax_a = plt.subplots()

ax_a.add_patch(patches.Rectangle((0.1,0.1),0.5,0.5, fill = False))
# Simple rectangle
```

```python
fig_b, ax_b = plt.subplots()

ax_b.add_patch(patches.Rectangle((0.1,0.1),0.5,0.5, fill = True))
# Color fIlling figure
```

```python
fig_c, ax_c = plt.subplots()

ax_c.add_patch(patches.Rectangle((0.1,0.1),0.5,0.5, facecolor = 'yellow', edgecolor = 'green'))
# Color fIlling figure
```

```python
# There is a way to create a plot adding several patches
fig_d, ax_d = plt.subplots()

ax_d.set_aspect(aspect = 'equal')

list_a = [
    patches.Rectangle((0.1,0.1),0.3,0.6, hatch = '.'),
    patches.Rectangle((0.5,0.1),0.3,0.6, hatch = '/', fill = False),
]

for patch in list_a:
    ax_d.add_patch(patch)

fig_d
```

```python
fig_e, ax_e = plt.subplots()

ax_e.set_aspect(aspect = 'equal')

list_b = [
    patches.Rectangle((0.1,0.1),0.2,0.6, alpha = None),
    patches.Rectangle((0.4,0.1),0.2,0.6, alpha = 1),
    patches.Rectangle((0.7,0.1),0.2,0.6, alpha = 0.6),
    patches.Rectangle((1,0.1),0.2,0.6, alpha = 0.1)
]

for patch in list_b:
    ax_e.add_patch(patch)

fig_e
# Even though there are 4 patches only 3 are displayed due to the fact that
# the 4th is outside the fig. To display it the axis must be extendend
```

```python
ax_e.set_xlim(0,1.3)
fig_e
# Tadah!
```

```python
# Diferent colors can be added
fig_f, ax_f = plt.subplots()

ax_f.set_aspect(aspect = 'equal')

list_c = [
    patches.Rectangle((0.1,0.1),0.2,0.6, facecolor = None), # Default color
    patches.Rectangle((0.4,0.1),0.2,0.6, facecolor = 'none'), # Literally no color
    patches.Rectangle((0.7,0.1),0.2,0.6, facecolor = 'red'),
    patches.Rectangle((1.0,0.1),0.2,0.6, facecolor = '#00ffff'),
]

for patch in list_c:
    ax_f.add_patch(patch)

ax_f.set_xlim(0,1.3)
fig_f
```

```python
# Different line colors can be added
fig_g, ax_g = plt.subplots()

ax_g.set_aspect(aspect = 'equal')

list_d = [
    patches.Rectangle((0.1,0.1),0.2,0.6, edgecolor = None, fill = False),
    patches.Rectangle((0.4,0.1),0.2,0.6, edgecolor = 'green', fill = False),
    patches.Rectangle((0.7,0.1),0.2,0.6, edgecolor = 'red', fill = False),
    patches.Rectangle((1.0,0.1),0.2,0.6, edgecolor = '#00ffff', fill = False),
]

for patch in list_d:
    ax_g.add_patch(patch)

ax_g.set_xlim(0,1.3)
fig_g
```

```python
# Different line width can be setted
fig_g, ax_g = plt.subplots()

ax_g.set_aspect(aspect = 'equal')

list_d = [
    patches.Rectangle((0.1,0.1),0.2,0.6, edgecolor = None, fill = False, linewidth = 0.5), 
    patches.Rectangle((0.4,0.1),0.2,0.6, edgecolor = 'gree', fill = False, linewidth = 1.5), 
    patches.Rectangle((0.7,0.1),0.2,0.6, edgecolor = 'red', fill = False, linewidth = 2.5), 
    patches.Rectangle((1.0,0.1),0.2,0.6, edgecolor = '#00ffff', fill = False, linewidth = 3.5), 
]

for patch in list_d:
    ax_g.add_patch(patch)

ax_g.set_xlim(0,1.3)
fig_g
```

```python
# Different line stile can be setted
fig_g, ax_g = plt.subplots()

ax_g.set_aspect(aspect = 'equal')

list_d = [
    patches.Rectangle((0.1,0.1),0.2,0.6, edgecolor = None, fill = False, linewidth = 0.5, linestyle = 'solid'), 
    patches.Rectangle((0.4,0.1),0.2,0.6, edgecolor = 'green', fill = False, linewidth = 1.5, linestyle = 'dashed'), 
    patches.Rectangle((0.7,0.1),0.2,0.6, edgecolor = 'red', fill = False, linewidth = 2.5, linestyle = 'dashdot'), 
    patches.Rectangle((1.0,0.1),0.2,0.6, edgecolor = '#00ffff', fill = False, linewidth = 3.5, linestyle = 'dotted'), 
]

for patch in list_d:
    ax_g.add_patch(patch)

ax_g.set_xlim(0,1.3)
fig_g
```

```python
fig_h, ax_h = plt.subplots()

ax_h.set_aspect(aspect = 'equal')

list_e = [
    patches.Arrow(0.2,0,0,0.5, edgecolor = None, facecolor = 'green', linewidth = 0.5, linestyle = 'solid'), 
    patches.Ellipse((0.4,0.1),0.1,1, edgecolor = 'green', facecolor = 'red', linewidth = 1.5, linestyle = 'dashed', hatch = '/'),  
    patches.Circle((0.6,0.1),0.1, edgecolor = 'red', facecolor = 'yellow', linewidth = 2.5, linestyle = 'dashdot', hatch = '**'), 
    patches.Arc((0.8,0.5),0.6,0.2, angle=90, theta1=90, theta2=270.0, edgecolor = '#00ffff', fill = False, linewidth = 3.5, linestyle = 'dotted', hatch = '\\'), 
    patches.FancyArrow(1,0,0,0.5, width = 0.05 ,edgecolor = '#0000ff', facecolor = 'blue', linewidth = 0.5, linestyle = 'solid', hatch = '//'), 
]

for patch in list_e:
    ax_h.add_patch(patch)

ax_h.set_xlim(0,1.3)
fig_h
# Welcome to the carnival
```

```python
# There's a way to plot polygons
fig_j, ax_j = plt.subplots()

ax_j.set_aspect(aspect = 'equal')

polygon_a = patches.Polygon(
    [[0.1,0.1],
    [0.2,0.8],
    [0.5,0.7],
    [0.8,0.1],
    [0.4,0.3]],
    fill = False
) # Polygon edges

ax_j.add_patch(polygon_a)

ax_j.set_xlim(0,1.3)
fig_j
# Closed polygon
```

```python
fig_j, ax_j = plt.subplots()

ax_j.set_aspect(aspect = 'equal')

polygon_a = patches.Polygon(
    [[0.1,0.1],
    [0.2,0.8],
    [0.5,0.7],
    [0.8,0.1],
    [0.4,0.3]],
    fill = False, closed = False
) # Polygon edges

ax_j.add_patch(polygon_a)

ax_j.set_xlim(0,1.3)
fig_j
# Open polygon
```

```python
# Way to draw arrows
fig_j, ax_j = plt.subplots()

ax_j.set_aspect(aspect = 'equal')

polygon_a = patches.Polygon(
    [[0.4,0.1],
    [0.4,0.5],
    [0.2,0.5],
    [0.5,0.8],
    [0.8,0.5],
    [0.6,0.5],
    [0.6,0.1]],
    fill = False
) # Polygon edges

ax_j.add_patch(polygon_a)

ax_j.set_xlim(0,1.3)
fig_j
# Open polygon
```
