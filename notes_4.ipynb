```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

matplotlib.interactive(True)
plt.ion()
matplotlib.is_interactive()
```

```python
# A figure in matplotlib is a high level container that
# represents the window or drawing area. This container
# recieves several plots on which all the X and Y will
# be drawn
fig_a = plt.figure()
ax_a = fig_a.add_axes([0,0,1,1])
# From: https://www.geeksforgeeks.org/how-to-add-axes-to-a-figure-in-matplotlib-with-python/

# Syntax: matplotlib.pyplot.figure.add_axes(rect)

# Parameters:

# rect: This parameter is the dimensions [xmin, ymin, dx, dy] of the new axes. It takes the below elements as arguments in the list:

# xmin: Horizontal coordinate of the lower left corner.
# ymin: Vertical coordinate of the lower left corner.
# dx: Width of the subplot.
# dy: Height of the subplot.
# Returns: This method return the axes class depends on the projection used.
```

```python
fig_b = plt.figure()
ax_b = fig_b.add_axes([0,0.6,0.9,0.4])
ax_c = fig_b.add_axes([0,0,0.7,0.4])
ax_d = fig_b.add_axes([0.9,0,0.6,0.4])
ax_e = fig_b.add_axes([1.1,0.6,0.4,0.4])
# Several axes are included in the 'fig_a' figure
```

```python
# Add data
x = np.linspace(start=0,stop=10,num=50)
ax_b.plot (x,np.sin(x))
ax_c.plot (x,np.cos(x))
ax_d.plot (x, x)
ax_e.plot (-x, x)
fig_b
```

```python
# Labeling
fontsize = 16
ax_b.set_xlabel ('X axis',fontsize = fontsize, color = 'r')
ax_b.set_ylabel ('Y axis',fontsize = fontsize, color = 'r')

ax_c.set_xlabel ('X axis',fontsize = fontsize, color = 'g')
ax_c.set_ylabel ('Y axis',fontsize = fontsize, color = 'g')

ax_d.set_xlabel ('X axis',fontsize = fontsize, color = 'y')
ax_d.set_ylabel ('Y axis',fontsize = fontsize, color = 'y')

ax_e.set_xlabel ('X axis',fontsize = fontsize, color = 'b')
ax_e.set_ylabel ('Y axis',fontsize = fontsize, color = 'b')
fig_b
```

```python
# Figure enlargement / resize
# The size is in inches
fig_b.set_size_inches(10, 10, forward=True)
fig_b
```

```python
fig_b.set_dpi(100)
fig_b
```

```python
# There is a way to set plot grids within a figure
fib_c = plt.figure(figsize = (8,8))
# Setting figure size from fig declaration
# To declare the ammount of subplots there has to be 
# passed an integer that contains the dimentions of the 
# grid. Inthis case 2x2. To define the axis we must provide
# a value. To define the first axis/subplot there has to
# be provided the 221 integer. For the second 222, and
# so on
ax_f = fib_c.add_subplot (221)
ax_g = fib_c.add_subplot (224)
```

```python
# Add data
ax_f.plot (x,np.sin(x))
ax_g.plot (x,np.cos(x))
fib_c
```

```python
# Labeling
fontsize = 16
ax_f.set_xlabel ('X axis',fontsize = fontsize, color = 'r')
ax_f.set_ylabel ('Y axis',fontsize = fontsize, color = 'r')

ax_g.set_xlabel ('X axis',fontsize = fontsize, color = 'g')
ax_g.set_ylabel ('Y axis',fontsize = fontsize, color = 'g')
fib_c
```

```python
# There's an additional way to create a figure with
# subplots
# subplot2grid recieves a 2D tuple with the shape of
# the grid and another 2D tuple for the coordinate
# of the axis

ax_h = plt.subplot2grid((2,3),(0,0))
ax_h.plot(x,np.sin(x))
ax_h.set_label('Sine Curve')

ax_i = plt.subplot2grid((2,3),(0,1))
ax_i.plot(x,np.cos(x))
ax_i.set_label('Cosine Curve')

ax_j = plt.subplot2grid((2,3),(0,2), rowspan = 2)
ax_j.plot(x,np.cos(x))
ax_j.set_label('Cosine Curve')
# This fix this overlapping issue there can be shifter
# the tick position of the plot
```

```python
ax_h = plt.subplot2grid((2,3),(0,0))
ax_h.plot(x,np.sin(x))
ax_h.set_label('Sine Curve')

ax_i = plt.subplot2grid((2,3),(0,1))
ax_i.plot(x,np.cos(x))
ax_i.set_label('Cosine Curve')

ax_j = plt.subplot2grid((2,3),(0,2), rowspan = 2)
ax_j.plot(x,np.cos(x))
ax_j.set_label('Cosine Curve')
ax_j.yaxis.tick_right()


ax_k = plt.subplot2grid((2,3),(1,0), colspan = 2)
ax_k.plot(x,np.exp(x))
ax_k.set_label('EXP')
```

```python
# Watermarks and text can be added to a plot

fig_e = plt.figure() # Fig
ax_m = fig_e.add_axes([0,0.6,0.9,0.4]) # Grid Layout
ax_n = fig_e.add_axes([0,0,0.7,0.4])
ax_o = fig_e.add_axes([0.9,0,0.6,0.4])
ax_p = fig_e.add_axes([1.1,0.6,0.4,0.4])

x = np.linspace(start=0,stop=10,num=50) # Values
ax_m.plot (x,np.sin(x))
ax_n.plot (x,np.cos(x))
ax_o.plot (x, x)
ax_p.plot (-x, x)

ax_m.text (1,-0.5, "Water mark 1", fontsize=32, color='r', ha="left", va="bottom", alpha=0.3)
ax_n.text (1,-0.5, "Water mark 2", fontsize=32, color='r', ha="left", va="bottom", alpha=0.3)
ax_o.text (1,3.5, "Water mark 3", fontsize=32, color='r', ha="left", va="bottom", alpha=0.3)
ax_p.text (-10,4, "Water mark 4", fontsize=32, color='r', ha="left", va="bottom", alpha=0.3)
fig_e

# Note that the 'text()' places text within the axis plot. This method recieves:
# 1 - The position on the X axis for the text
# 2 - The position on the Y axis for the text
# 3 - The strinh to display
# 4 - The font size
# 5 - Color
# 6 - Horizontal align
# 7 - Vertical align
# 8 - Transparency
```
