```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

matplotlib.interactive(True)
plt.ion()
matplotlib.is_interactive()
```

```python
# There can be created a plot with multiple lines
x = np.linspace(start = 0, stop = 10, num = 50)
plt.plot(x,np.sin(x), label = 'Sine Curve')
plt.plot(x,np.cos(x), label = 'Cosine Curve')
plt.legend()
plt.title ('Trigonometric Plot')
```

```python
# These lines can be modded
# Line colors can be setted with the starting letter of 
# a color
plt.plot(x,np.sin(x), label = 'Sine Curve', color = 'y')
plt.plot(x,np.cos(x), label = 'Cosine Curve', color = 'g')
plt.legend()
plt.title ('Trigonometric Plot')
```

```python
# There can be modified the type of line
array_a = np.random.randn(20)
plt.plot(array_a, color ='m', linestyle = ':')
```

```python
# Different lines
plt.plot(array_a, color ='m', linestyle = '--')
```

```python
# Different line thiccness
plt.plot(array_a, color ='m', linestyle = ':', linewidth = 3)
```

```python
# There can be displayed the actual values from a line
plt.plot(array_a, color ='m', linestyle = ':', marker = 'd')
```

```python
# Such markers can be resized
plt.plot(array_a, color ='m', linestyle = ':', marker = 'd', markersize = 10)
```

```python
# Line can be removed
plt.plot(array_a, color ='m', linestyle = 'None', marker = 'd')
```

```python
# There can be plotted the relation between two variables
plt.scatter(np.random.randn(20),array_a, color = 'g')
plt.scatter(np.random.randn(20),array_a, color = 'y')
```
