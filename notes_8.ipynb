```python
import matplotlib
import matplotlib.pyplot as plt

matplotlib.interactive(True)
plt.ion()
matplotlib.is_interactive()
```

```python
# Any plot can store anotations. This are setted up the following manner

fig_a , ax_a = plt.subplots()

ax_a. plot([1,2,3],
            [2,4,6])

ax_a.annotate('Min Value',xy= (1,2), xytext = (1.5,2.0), arrowprops = dict(color='g'))
```

```python
# To change arrows direction there can be modified the 'xytext'
fig_b , ax_b = plt.subplots()

ax_b. plot([1,2,3],
            [2,4,6])

ax_b.annotate('Min Value',xy= (1,2), xytext = (1,3), arrowprops = dict(color='g'))
```

```python
# There can be moddified the arrows properties
# To change arrows direction there can be modified the 'xytext'
fig_c , ax_c = plt.subplots()

ax_c. plot([1,2,3],
            [2,4,6])

ax_c.annotate('Min Value',xy= (1,2), xytext = (1,3), arrowprops = dict(facecolor='b',edgecolor='y',alpha=0.5))
```

```python
# There can be added aditional notations to the plot
fig_d , ax_d = plt.subplots()

ax_d. plot([1,2,3],
            [2,4,6])

ax_d.annotate('Min Value',xy= (1.75,3.5), xytext = (1.75,4.5), arrowprops = dict(facecolor='b',edgecolor='y',alpha=0.5), color='orange', style='italic')
ax_d.plot([1.75],[3.5],'ro')
```

```python
# To avoid overlapping as the figure shown above some objects can be shrink
fig_e , ax_e = plt.subplots()

ax_e. plot([1,2,3],
            [2,4,6])

ax_e.annotate('Min Value',xy= (1.75,3.5), xytext = (1.75,4.5), arrowprops = dict(facecolor='b',edgecolor='y',alpha=0.5, shrink=0.1), color='orange', style='italic')
ax_e.plot([1.75],[3.5],'ro')
```

```python
# Advanced anotations
import numpy as np

#Plot data
np.random.seed(0)
# Skew bottom left
x1 = - 1 + np.random.randn(100)
y1 = - 1 + np.random.randn(100)

# Skew top right
x2 = 1 + np.random.randn(100)
y2 = 1 + np.random.randn(100)

# Plot
fig_f, ax_f = plt.subplots()
ax_f.scatter(x1,y1, color='b')
ax_f.scatter(x2,y2, color='y')
```

```python
# Bounding box
# Simple boxes
b_box_a = dict(boxstyle = 'square', facecolor = 'w', alpha=0.3)

ax_f.text(-2,-2, 'Sample A', ha='center',va='center', size=20, bbox = b_box_a)
ax_f.text(2,2, 'Sample B', ha='center',va='center', size=20, bbox = b_box_a)
fig_f
```

```python
# Arrows
b_box_b = dict(boxstyle = 'rarrow', facecolor = '#EBF5FB', edgecolor = 'b', linewidth = 2, alpha=0.3)
b_box_c = dict(boxstyle = 'larrow', facecolor = '#EBF5FB', edgecolor = 'b', linewidth = 2, alpha=0.3)
b_box_d = dict(boxstyle = 'darrow', facecolor = '#EBF5FB', edgecolor = 'b', linewidth = 2, alpha=0.3)

ax_f.text(-2.5,1, 'Sample C', ha='center',va='center', rotation = 90, size=20, bbox = b_box_b)
ax_f.text(2.5,-1, 'Sample D', ha='center',va='center', rotation = 90, size=20, bbox = b_box_c)
ax_f.text(0,0, 'Sample E', ha='center',va='center', rotation = 60, size=20, bbox = b_box_d)
fig_f
```

```python
# Moddified boxes
b_box_e = dict(boxstyle = 'sawtooth', facecolor = '#EBF5FB', edgecolor = 'b', linewidth = 2, alpha=0.3)
b_box_f = dict(boxstyle = 'round4', facecolor = '#EBF5FB', edgecolor = 'b', linewidth = 2, alpha=0.3)

ax_f.text(-1,1.5, 'Sample F', ha='center',va='center', size=10, bbox = b_box_e)
ax_f.text(1,-1.5, 'Sample G', ha='center',va='center', size=10, bbox = b_box_f)
fig_f
```
